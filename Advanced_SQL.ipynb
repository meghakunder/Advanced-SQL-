**Q1. What is a Common Table Expression (CTE), and how does it improve SQL query readability?**

**Ans:**

A Common Table Expression (CTE) is a temporary, named result set defined within a SQL statement using the WITH clause. It exists only for the duration of the query and can be referenced like a table or view in the main query.

WITH cte_name AS  
(

    SELECT column1, column2

    FROM table_name

    WHERE condition
)

SELECT *

FROM cte_name;

It is evaluated at runtime and exists only for the scope of the executing statement. Unlike derived tables or nested subqueries, a CTE allows complex query logic to be expressed in a modular and readable form.

**CTEs improve SQL readability** by separating intermediate computations from the final query logic. This enables a top-down flow of execution, where each logical step is clearly named and defined, reducing cognitive load and minimizing nested query complexity. As a result, queries become easier to debug, maintain, and extend.

**Q2. Why are some views updatable while others are read-only? Explain with an example.**

**Ans:**

Some views are updatable while others are read-only because it depends on how the view is defined and whether the database can clearly map changes in the view back to rows in the base table.

* A view is updatable if it is based on a single base table.
* The view must not contain JOIN, GROUP BY, DISTINCT, aggregate functions (SUM, COUNT, etc.), or set operations.
* Each row in the view must correspond to one row in the base table.
* A view becomes read-only if it:
  * Uses multiple tables (JOIN)
  * Contains aggregate functions
  * Uses GROUP BY, HAVING, or DISTINCT
  * Includes calculated or derived columns

**Example**

**Updatable View**

CREATE VIEW active_employees AS

SELECT emp_id, name, salary

FROM employees

WHERE status = 'Active';

This view is updatable because it is based on a single table and has no aggregations.

**Read-Only View**

CREATE VIEW dept_salary AS

SELECT department_id, AVG(salary) AS avg_salary

FROM employees

GROUP BY department_id;

This view is read-only because it uses an aggregate function and GROUP BY, so updates cannot be mapped to individual rows.

**Q3. What advantages do stored procedures offer compared to writing raw SQL queries repeatedly?**

**Ans:**

* Stored procedures allow reusability of SQL logic, avoiding repeated query writing.

* They improve performance by reducing query compilation and execution overhead.

* Stored procedures enhance security by allowing controlled access to data without exposing underlying tables.

* They make applications easier to maintain, since changes are made in one place.

* They reduce network traffic, as only the procedure call is sent instead of full SQL statements.

**Q4. What is the purpose of triggers in a database? Mention one use case where a trigger is essential.**

**Ans:**

* Triggers are database objects that automatically execute when a specific event occurs on a table (INSERT, UPDATE, or DELETE).

* They are used to enforce business rules, maintain data integrity, and perform automatic actions.

* Triggers run without user intervention, ensuring consistency at the database level.

Use case:

Automatically updating an audit or log table whenever a record is updated or deleted, to track changes for security and compliance purposes.

**Q5. Explain the need for data modelling and normalization when designing a database.**

**Ans:**

* Data modelling helps define the structure of the database by identifying entities, attributes, and relationships.

* It ensures the database design reflects business requirements accurately.

* Proper data modelling improves data consistency and scalability.

* Normalization organizes data to eliminate redundancy.

* It prevents data anomalies during insert, update, and delete operations.

* Normalization improves data integrity and makes the database easier to maintain.
